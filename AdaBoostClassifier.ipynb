```python
# Load libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
# Load data
#iris = datasets.load_iris()

import pandas as pd
#from sklearn import datasets

dataset = pd.read_csv("bitcoin_int.csv")
#dataset = pd.read_csv("smite2_items.csv")
#dataset = datasets.load_iris()
dataset.shape
```

```python
# bit coin dataset
x = dataset.drop(['Direction','WeekYear'], axis=1)
y = dataset['Direction']  

# smite dataset
#x = dataset.drop(['Name','Tier','Passive','Active'], axis=1)
#y = dataset['Tier']  
```

##### Split the dataset

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2) # 70% training and 30% test
```

## 1) base_estimator = Decision Tree Classifier 

```python
# Create adaboost classifer object
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
```

- Nếu không khai báo thì AdaBoost mặc định sử dụng  Decision Tree Classifier 

```python
abc
```

- base_estimator: weak learner
- n_estimators: number  of weak learners to train iteratively.


```python
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)
```

```python
model
```

```python


#Predict the response for test dataset
y_pred = model.predict(X_test)
```

```python
y_pred
```

##### Evaluation

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

## 2) Using Different Base Learners
- base_estimator = Support Vector Classifier 

```python
# Import Support Vector Classifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')
```

```python
# Create adaboost classifer object
abc =AdaBoostClassifier(n_estimators=50, estimator=svc,learning_rate=1)
```

```python
abc
```

```python
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)
model
```

```python
#Predict the response for test dataset
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

- Có thể thấy khi sử dụng Support Vector Classifier(SVC) làm weak learners độ chính xác tăng lên
